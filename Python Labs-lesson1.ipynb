```python
import numpy as np
import matplotlib.pyplot as plt
# Create an array of x values
x = np.linspace(0, 10, 100)
# Compute y values
y = np.sin(x)
# Plot the results
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.show()
```

```python
from geopy.geocoders import Nominatim
from geopy import distance

# 初始化地理编码器
geolocator = Nominatim(user_agent="my_geocoder")

# 获取两个地点的坐标
location1 = geolocator.geocode("helsinki")
location2 = geolocator.geocode("espoo")

# 计算两地之间的直线距离（单位：公里）
print(distance.distance((location1.latitude, location1.longitude),
                        (location2.latitude, location2.longitude)).kilometers)
```
